```python
#importing pandas
import pandas as pd
import numpy as np
```

```python
#importing nfl_csv
nfl_df = pd.read_csv('nfl.csv')

nfl_df
```

```python
draft_df = nfl_df['Drafted..tm.rnd.yr.'].str.split('/', expand = True)
draft_df
```

```python
#rename drafted_df columns
# needs to be added into main DF
draft_df.columns = ['Team', 'Round', 'Overall_Pick', 'Year']
draft_df
```

```python
#set column from nfl_df to variable
Player = nfl_df['Player']
```

```python
#add player column from nfl_df to drafted_df
draft_df = draft_df.join(Player)
draft_df
```

```python
#drop team column
draft_df.drop(columns = ['Team'], inplace = True)
draft_df
```

```python
#adding columns to draft_df
draft_df['Sprint_40yd'] = nfl_df['Sprint_40yd']
draft_df['Vertical_Jump'] = nfl_df['Vertical_Jump']
draft_df['Bench_Press_Reps'] = nfl_df['Bench_Press_Reps']
draft_df['Broad_Jump'] = nfl_df['Broad_Jump']
draft_df['Agility_3cone'] = nfl_df['Agility_3cone']
draft_df['Shuttle'] = nfl_df['Shuttle']
draft_df['Position_Type'] = nfl_df['Position_Type']
draft_df['Drafted'] = nfl_df['Drafted']
```

```python
draft_df.head()
```

```python
#reorder columns
combine_df = draft_df.loc[:,['Year', 'Player', 'Position_Type', 'Round', 'Overall_Pick', 'Sprint_40yd', 'Vertical_Jump', 'Bench_Press_Reps', 'Broad_Jump', 'Agility_3cone', 'Shuttle', 'Drafted']]
```

```python
#stats for drafted players
players_drafted_df = combine_df.loc[draft_df['Drafted'] == 'Yes']
players_drafted_df
```

```python
#stats for undrafted players
players_undrafted_df = combine_df.loc[draft_df['Drafted'] == 'No']
players_undrafted_df
```
