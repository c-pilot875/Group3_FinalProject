```python
#import pandas
import pandas as pd
import numpy as np
```

```python
#import dataset
nfl_df = pd.read_csv('nfl.csv')
```

```python
#show dataset
nfl_df
```

```python
#separate Drafted..tm.rnd.yr. column
draft_df = nfl_df['Drafted..tm.rnd.yr.'].str.split('/', expand = True)
draft_df
```

```python
#rename drafted_df columns
# needs to be added into main DF
draft_df.columns = ['Team', 'Round', 'Overall_Pick', 'Year']
draft_df
```

```python
#set column from nfl_df to variable
Player = nfl_df['Player']
```

```python
#add player column from nfl_df to drafted_df
combine_df = draft_df.join(Player)
combine_df
```

```python
combine_df.loc[:,['Year', 'Player', 'Team', 'Round', 'Overall_Pick']]
```

```python
# add school
combine_df['School'] = nfl_df['School']
combine_df
```

```python
#reorder updated combine_df
combine_df = combine_df.loc[:,['Year', 'Player', 'School', 'Round', 'Overall_Pick', 'Team']]
combine_df
```

```python
#Add drafted column from nfl_df to combine_df
combine_df['Drafted'] = nfl_df['Drafted']
combine_df
```

```python
#
players_drafted_df = combine_df.loc[combine_df['Drafted'] == 'Yes']
players_drafted_df
```

```python
#Sort drafted players df by School
players_drafted_df.sort_values(by=['School'])
```

```python
players_drafted_df['Sprint_40yd'] = nfl_df['Sprint_40yd']
players_drafted_df
```

```python
players_drafted_df['Vertical_Jump'] = nfl_df['Vertical_Jump']
players_drafted_df
```

```python
players_drafted_df['Bench_Press_Reps'] = nfl_df['Bench_Press_Reps']
players_drafted_df
```

```python
players_drafted_df['Broad_Jump'] = nfl_df['Broad_Jump']
players_drafted_df
```

```python
players_drafted_df['Agility_3cone'] = nfl_df['Agility_3cone']
players_drafted_df
```

```python
players_drafted_df['Shuttle'] = nfl_df['Shuttle']
players_drafted_df
```

```python
#Count for each school for drafted players
players_drafted_df['School'].value_counts()
```

```python
#add drafted result column to undrafted players df
players_undrafted_df = combine_df.loc[combine_df['Drafted'] == 'No']
players_undrafted_df
```

```python
#sort undrafted players by school
players_undrafted_df.sort_values(by=['School'])
```

```python
#Count of each school for undrafted players
players_undrafted_df['School'].value_counts()
```

```python
#Delete unnecessary columns from undrafted players df
del players_undrafted_df['Round']
del players_undrafted_df['Overall_Pick']
del players_undrafted_df['Team']
```

```python
players_undrafted_df
```

```python
# export CSVs
players_drafted_df.to_csv('Drafted_Players')
players_undrafted_df.to_csv('Undrafted_players')
```
