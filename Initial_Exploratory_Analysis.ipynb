```python
# Import Dependencies
import pandas as pd
```

```python
# Look at data
nfl_df = pd.read_csv("NFL.csv")
nfl_df.head()
```

```python
# Get rid of "Drafted..tm.rnd.yr" column since it redundant to "Drafted" column for initial purposes
nfl_df = nfl_df.drop("Drafted..tm.rnd.yr.", axis=1)
```

```python
# Check to see what other columns have null values
nfl_df.info()
```

```python
# Look more specifically at players that have null values for age
nfl_df.loc[pd.isnull(nfl_df['Age'])]
```

```python
# Drop the age column to increase number of rows, since we think maybe age isn't as important
nfl_df = nfl_df.drop("Age", axis=1)
```

```python
# Drop all other null values and see that we have 1731 data points remaining
nfl_df = nfl_df.dropna()
nfl_df.info()
```
