```python
import pandas as pd
```

```python
# Read the NFL.csv into a Dataframe
nfl_df = pd.read_csv('NFL.csv')
nfl_df
```

```python
# Split the Drafted..tm.rnd.yr. column and update headers
nfl_df[['Team', 'Round', 'Overall_Pick', 'Draft_Year']] = nfl_df['Drafted..tm.rnd.yr.'].str.split(' / ', expand = True)
nfl_df.head()
```

```python
# Drop original Drafted..tm.rnd.yr. column
nfl_df = nfl_df.drop('Drafted..tm.rnd.yr.',axis=1)
nfl_df.head()
```

```python
# Reorder the columns
new_column_order = ['Year','Player','Age','School','Height','Weight','Sprint_40yd','Vertical_Jump','Bench_Press_Reps'
                   ,'Broad_Jump','Agility_3cone','Shuttle','Team','Round','Overall_Pick','Draft_Year','BMI'
                   ,'Player_Type','Position_Type','Position','Drafted']

# Recreate the DataFrame by using the new column order
nfl_df = nfl_df[new_column_order]
nfl_df
```

```python
# Cleanup duplicated schools
nfl_df['School'] = nfl_df['School'].str.replace('LSU','Louisiana St')
nfl_df['School'] = nfl_df['School'].str.replace('West. Michigan','Western Michigan')
```

```python
# Export nfl_df
output_data_file = "NFL_Clean_for_DB.csv"
nfl_df.to_csv(output_data_file, index = False)
```
